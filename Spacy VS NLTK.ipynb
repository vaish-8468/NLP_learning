```python
import spacy
```

Sentence tokenization in spacy(using object oriented programming)

```python
nlp=spacy.load("en_core_web_sm")
doc=nlp("Dr. Vasu is a vety good boy. He completes his homework on time.")
for sentence in doc.sents:
  print(sentence)
```

```python
for sentence in doc.sents:
  for words in sentence:
    print(words)
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize("Dr. Vasu is a very good boy. He completes his homework on time.")
```
